```python
import pytesseract
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/Example05.png")
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

options = "outputbase digits"

text = pytesseract.image_to_string(rgb, config="outputbase digits")
print(text)
```

```python
h, w, _ = image.shape
boxes = pytesseract.image_to_boxes(rgb, config="outputbase digits")
print(boxes)
```

```python
plt.figure(figsize=(10, 10))
for b in boxes.splitlines():
	b = b.split(" ")
	x, y, x2, y2 = int(b[1]), int(b[2]), int(b[3]), int(b[4])
	cv2.rectangle(image, (x, h - y), (x2, h - y2), (0, 0, 255), 2)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
